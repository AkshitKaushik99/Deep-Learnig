# thresholding
<b>cv2.THRESH_BINARY_INV</b><br>
Converts pixels below t to 255(white) and above t to 0(black)

```python
_,t1 = cv2.threshold(rg,120,255,cv2.THRESH_BINARY)
plt.imshow(t1,'gray')
plt.show()
```

```python
k = np.ones((5,5))
er1 = cv2.erode(t1,k,iterations =3)
dili = cv2.dilate(t1,k,iterations =5)

# plt.imshow(rg,'gray'),plt.title('orignal'),plt.show()
plt.imshow(er1,'gray'),plt.title('Erodded'),plt.show()
# plt.imshow(dili,'gray'),plt.title('Dilated'),plt.show()
```

```python
contours , hier = cv2.findContours(er1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
rgb = cv2.cvtColor(er1,cv2.COLOR_BGR2RGB)
res = cv2.drawContours(rg,contours,-1,(0,0,255),2)
plt.imshow(res)
plt.show()
```

```python
r1 = cv2.imread('broken_grain_1.jpg')
plt.imshow(r1)
```

```python

def Object_Counting(thersholding,i,w,b):
    rg =cv2.cvtColor(r1,cv2.COLOR_BGR2GRAY)
    plt.imshow(rg,'gray')
    plt.axis('off')
    plt.show()
    _,t1 = cv2.threshold(rg,w,b,thersholding)
    plt.imshow(t1,'gray')
    plt.show()
    k = np.ones((5,5))
    er = cv2.erode(t1,k,iterations = i)
    plt.imshow(er1,'gray'),plt.title('Erodded'),plt.show()
    contours , hier = cv2.findContours(er,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    rgb = cv2.cvtColor(er,cv2.COLOR_BGR2RGB)
    res = cv2.drawContours(rg,contours,-1,(0,0,255),2)
    plt.imshow(res)
    plt.show()
    res1 = cv2.drawContours(r1,contours,-1,(0,0,255),2)
    plt.imshow(res1)
    plt.show()
    c = 0
    for i in contours:
        c +=1
    print(c)
```

```python
Object_Counting(cv2.THRESH_BINARY,2,120,255)
```
